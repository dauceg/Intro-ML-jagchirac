```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from   sklearn import linear_model
from   sklearn.metrics import mean_squared_error, r2_score
import math
from math import *
#Method : closed-form solution 
N_TData = 900

ID  = np.zeros(N_TData)
y   = np.zeros(N_TData)
x1  = np.zeros(N_TData)
x2  = np.zeros(N_TData)
x3  = np.zeros(N_TData)
x4  = np.zeros(N_TData)
x5  = np.zeros(N_TData)
x6  = np.zeros(N_TData)
x7  = np.zeros(N_TData)
x8  = np.zeros(N_TData)
x9  = np.zeros(N_TData)
x10 = np.zeros(N_TData)


# Reading the training data
# row = ligne,  for row in reader row['Id'] = à la ligne ou on est, la valeur qui correspond à Id .
with open('Train1B.csv') as csvfile:
    reader = csv.DictReader(csvfile, delimiter = ',')
    for row in reader : 
        key = int(row['Id'])
        ID[key] = row['Id'] 
        y[key]  = row['y']
        x1[key] = row['x1']
        x2[key] = row['x2']
        x3[key] = row['x3']
        x4[key] = row['x4']
        x5[key] = row['x5']
        
# phi functions implementation
def square(x):
    results = []
    for i in range(N_TData):
        results += [x[i]*x[i]]
    return(results)

def exp(x):
    results =[]
    for i in range(N_TData):
        results += [math.exp(x[i])]
    return(results)

def cos(x):
    results = []
    for i in range(N_TData):
        results += [math.cos(x[i])]
    return(results)

# X matrix implementation 
Xtrain = (np.vstack((x1, x2, x3, x4, x5))).T  

X = np.concatenate((Xtrain, np.array([square(x1)]).T), axis = 1)            
X = np.concatenate((X, np.array([square(x2)]).T), axis = 1)
X = np.concatenate((X, np.array([square(x3)]).T), axis = 1)
X = np.concatenate((X, np.array([square(x4)]).T), axis = 1)
X = np.concatenate((X, np.array([square(x5)]).T), axis = 1)

X = np.concatenate((X, np.array([exp(x1)]).T), axis = 1)            
X = np.concatenate((X, np.array([exp(x2)]).T), axis = 1)
X = np.concatenate((X, np.array([exp(x3)]).T), axis = 1)
X = np.concatenate((X, np.array([exp(x4)]).T), axis = 1)
X = np.concatenate((X, np.array([exp(x5)]).T), axis = 1)

X = np.concatenate((X, np.array([cos(x1)]).T), axis = 1)            
X = np.concatenate((X, np.array([cos(x2)]).T), axis = 1)
X = np.concatenate((X, np.array([cos(x3)]).T), axis = 1)
X = np.concatenate((X, np.array([cos(x4)]).T), axis = 1)
X = np.concatenate((X, np.array([cos(x5)]).T), axis = 1)

X = np.concatenate((X, np.ones((N_TData, 1))), axis = 1)

w = np.dot(np.linalg.pinv(np.dot(X.T, X)), np.dot(X.T, y))


# ----- WRITE DATA INTO CSV FILE


f = open('results1B.csv', 'w')

with f:

    writer = csv.writer(f)
    
    #for row in np.transpose(SaveData):
    writer.writerows([[hit] for hit in w])
```
